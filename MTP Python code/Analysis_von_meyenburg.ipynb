```python
# Now, we can use pandas read_excel after installing the excel importer.

import pandas as pd

df = pd.read_excel('data_von_meyenburg.xlsx')
df
```

```python
sf = 28
Mw_biomass = 25.17
```

```python

col = df.columns
row = df.index

q_table = pd.DataFrame()

q_table["D"] = df[col[0]]

q_table["ùíí_ùíô (C mole / L.h)"] = (df[col[1]]*0.92/Mw_biomass) * df[col[0]]
q_table["‚àíùíí_ùíî (C mole / L.h)"] = (28 - df[col[2]]) * df[col[0]]/30

q_table["ùíí_ùíë (C mole / L.h)"] = df[col[3]] * df[col[0]]/23
q_table["ùíí_ùíÑ (C mole / L.h)"] = df[col[4]] * df[col[1]] * 0.001
q_table["ùíí_ùíê (mole / L.h)"] = df[col[5]] * df[col[1]] * 0.001
q_table
```

```python
col = df.columns
row = df.index

df2 = pd.DataFrame()

df2["D"] = df[col[0]]
df2["DW"] = df[col[1]]

df2['CH2O'] = 1.00
df2['O2'] = q_table["ùíí_ùíê (mole / L.h)"] / q_table["‚àíùíí_ùíî (C mole / L.h)"]
df2['NH3'] = None
df2['CH1.83O0.56N0.17'] = q_table["ùíí_ùíô (C mole / L.h)"] / q_table["‚àíùíí_ùíî (C mole / L.h)"]
df2['H2O'] = None
df2['CO2'] = q_table["ùíí_ùíÑ (C mole / L.h)"] / q_table["‚àíùíí_ùíî (C mole / L.h)"]
df2['CH3O0.5'] = q_table["ùíí_ùíë (C mole / L.h)"] / q_table["‚àíùíí_ùíî (C mole / L.h)"]
df2
```

```python
reactants = ['CH2O', 'O2', 'NH3']
products = ['CH1.83O0.56N0.17', 'H2O', 'CO2', 'CH3O0.5']

error_percent_orig = [5, 10, None, 5, None, 10, 5]

analysis1 = Analysis(reactants, products, error_percent_orig, df2)
analysis1.perform_analysis()
```
